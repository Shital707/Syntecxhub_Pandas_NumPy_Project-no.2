## PROJECT 1: Time Series & Category Charts

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("sales_data.csv")

# Convert date column
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])

#1. Sales Over Time (Line Chart)
plt.figure()
plt.plot(df['Sale_Date'], df['Sales_Amount'])
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("sales_over_time.png")
plt.show()
```

```python
# 2. Monthly Sales Aggregation
df['Month'] = df['Sale_Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales_Amount'].sum()

plt.figure()
monthly_sales.plot()
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 3. Sales by Product Category (Bar Chart)
category_sales = df.groupby('Product_Category')['Sales_Amount'].sum()

plt.figure()
category_sales.plot(kind='bar')
plt.title("Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.savefig("category_bar_chart.png")
plt.show()
```

```python
# 4. Category Share (Pie Chart)
plt.figure()
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Category Sales Share")
plt.ylabel("")
plt.tight_layout()
plt.savefig("category_pie_chart.png")
plt.show()
```

```python
# 5. Export Summary
with open("summary.txt", "w") as f:
    f.write("SALES SUMMARY REPORT\n")
    f.write("=====================\n")
    f.write(f"Total Sales: {df['Sales_Amount'].sum()}\n")
    f.write(f"Highest Sales Category: {category_sales.idxmax()}\n")
    f.write(f"Best Performing Region: {df.groupby('Region')['Sales_Amount'].sum().idxmax()}\n")

print("Project 1 Completed Successfully!")
```
